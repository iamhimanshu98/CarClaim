```python
df.replace('?', np.nan, inplace=True)
```

```python
df.isna().sum()
```

```python
df['collision_type'] = df['collision_type'].fillna(df['collision_type'].mode()[0])
df['property_damage'] = df['property_damage'].fillna(df['property_damage'].mode()[0])
df['police_report_available'] = df['police_report_available'].fillna(df['police_report_available'].mode()[0])
df['authorities_contacted'] = df['authorities_contacted'].fillna(df['authorities_contacted'].mode()[0])
```

```python
df.isna().sum()
```

```python
# Assuming your original DataFrame is named 'original_df'
selected_columns = ['age', 'insured_sex', 'insured_education_level', 'policy_annual_premium', 'policy_deductable', 'incident_type', 'collision_type', 'incident_severity', 'authorities_contacted', 'number_of_vehicles_involved', 'property_damage', 'bodily_injuries', 'witnesses', 'injury_claim', 'vehicle_claim', 'fraud_reported']

refined_df = df[selected_columns]
```

```python
refined_df.head(10)
```

```python
df['insured_education_level'].value_counts()
```

```python
refined_df['insured_education_level'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
refined_df['insured_sex'] = lb.fit_transform(refined_df['insured_sex'])
refined_df['insured_education_level'] = lb.fit_transform(refined_df['insured_education_level'])
refined_df['incident_type'] = lb.fit_transform(refined_df['incident_type'])
refined_df['collision_type'] = lb.fit_transform(refined_df['collision_type'])
refined_df['incident_severity'] = lb.fit_transform(refined_df['incident_severity'])
refined_df['authorities_contacted'] = lb.fit_transform(refined_df['authorities_contacted'])
refined_df['property_damage'] = lb.fit_transform(refined_df['property_damage'])
```

```python
import joblib

filename = 'label_encoder.pkl'
joblib.dump(lb, filename)
```

```python
refined_df.head(2)
```

```python
x = refined_df
x = x.drop(['fraud_reported'], axis=1)
x.head()
y = refined_df['fraud_reported']
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=40, test_size=0.25)
```

```python
refined_df.head(2)
```

```python
refined_df['number_of_vehicles_involved'].value_counts()
```

```python
dtc = DecisionTreeClassifier()
dtc = dtc.fit(x_train,y_train)

print("Model trained Successfully")
dtc.score(x_train,y_train) * 100
dtc.score(x_test,y_test) * 100
```

```python
import joblib, os
os.makedirs('model',exist_ok=True)
```

```python
joblib.dump(dtc, "./model/decision_tree_model.lb")
print("Model saved successfully")
```

```python
prediction = dtc.predict(x_test)
```

```python
compare_df = pd.concat([x_test, y_test],axis='columns')
compare_df['prediction'] = prediction
compare_df.head(10)
compare_df.shape
```

```python
os.makedirs("prediction", exist_ok=True)
```

```python
compare_df.to_csv('./prediction/prediction.csv', index=False)
print("Result saved successfully")
```
